```python
import pandas as pd
from IPython.display import display

class CabanaMontana:
    def __init__(self, nume_cabana, pret_noapte):
        self.nume = nume_cabana
        self.pret_noapte = pret_noapte
        self.registru = []
    def checkin(self,nume,varsta,nivel_experienta,nopti_cazare):
        turist={
           "Nume":nume,
           "Varsta":varsta,
           "Experienta":nivel_experienta,
           "Nopti cazare":nopti_cazare
        }
        self.registru.append(turist)
        print(f"{nume} a facut check-in la {self.nume} pentru {nopti_cazare} nopti.")
    def genereaza_raport(self):
        if len(self.registru) == 0:
            print(f"Cabana {self.nume} este goala momentan.")
            return
        df=pd.DataFrame(self.registru)
        df['Nota_Plata']=df['Nopti cazare']*self.pret_noapte
        display(df)
        incasari_totale = df['Nota_Plata'].sum()
        print(f"\n Incasari totale: {incasari_totale} RON")

        print("\n Tipuri de turisti cazati (groupby):")
        statistici_experienta = df.groupby('Experienta')['Nume'].count()
        print(statistici_experienta)
cabanaomu=CabanaMontana("Cabana Omu",150)
cabanaomu.checkin("Tista Bogdan",20,"ridicat",4)
cabanaomu.checkin("Covasan Flavius",21,"modest",1)
cabanaomu.checkin("Gubernat Adrian",5,"modest",7)
cabanaomu.genereaza_raport()
```
