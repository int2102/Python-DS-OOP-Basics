```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

experiment_efhall = {
    "B(mT)":[0,50,100,150,200,250,300],
    "I=10mA":[8.3,12.8,15.9,21.6,27.8,28.3,30.5],
    "I=20mA":[10,20.1,26,32.4,39.5,46.7,56.8],
    "I=30mA":[11.7,27.2,34.1,49.5,57.8,69.3,78.1]
}

df = pd.DataFrame(experiment_efhall)
df['Bmean']=(df['I=10mA']+df['I=20mA']+df['I=30mA'])/3
df.loc['Mean']=df[['I=10mA','I=20mA','I=30mA','Bmean']].mean()
df.loc['Mean']=df.loc['Mean'].fillna(0)
display(df)
sns.set_theme(style="whitegrid")
df_grafic=df.iloc[0:7,0:4]
sns.regplot(data=df_grafic,x="I=10mA",y="B(mT)",color="purple",label="I=10mA")
sns.regplot(data=df_grafic,x="I=20mA",y="B(mT)",color="red",label="I=20mA")
sns.regplot(data=df_grafic,x="I=30mA",y="B(mT)",color="yellow",label="I=30mA")
plt.title("Efectul Hall - Regresie Liniară Multiplă")
plt.legend()
plt.show()
```
